## Segment 3 - Data parsing

```python
from bs4 import BeautifulSoup

import urllib
import urllib.request
import re

from IPython.display import HTML
```

```python
with urllib.request.urlopen('https://stackoverflow.com/questions/27243129/how-to-open-html-file.html') as response:
    html = response.read()
```

```python
soup = BeautifulSoup(html, "lxml")
type(soup)
```

### Parsing your data

```python
print(soup.prettify()[0:500]) # must use print to make each tag in single line
```

### Searching and retrieving data from a parse tree

#### Retrieving tags by filtering with name arguments

```python
soup.find_all("li")
```

##### Retrieving tags by filtering with keyword arguments

```python
soup.find_all(id="class") # 需要在tag中明确identify id=...才能搜索到
```

##### Retrieving tags by filtering with string arguments

```python
soup.find_all('li')  #  不太懂这个和第一个有啥区别，引号？
```

##### Retrieving tags by filtering with list objects

```python
soup.find_all(['ol','a'])
```

##### Retrieving tags by filtering with regular expressions

```python
l = re.compile("l")
for tag in soup.find_all(l):
    print(tag.name)
```

##### Retrieving tags by filtering with a Boolean value

```python
for tag in soup.find_all(True): # 输入false结果一样的啊？
    print(tag.name)
```

##### Retrieving weblinks by filtering with string objects

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```

##### Retrieving strings by filtering with regular expressions

```python

soup.find_all(string=re.compile("Stack"))
```

### Segment 4 - Web Scraping


```python
r = urllib.request.urlopen('https://analytics.usa.gov/').read()
soup = BeautifulSoup(r, "lxml")
type(soup)
```

```python
print(soup.prettify()[0:200])
```

```python
for tag in soup.find_all('a', attrs={'href': re.compile("http")}):  
    print(tag)  # type of the link is tag

# it does work when change the tag to link: for link in ... print(link)； type is all the same
```

```python
file = open("parsed_data.txt", "w") # w for write
for link in soup.find_all('a', attrs={'href': re.compile("http")}):  
    print(link)
    file.write(link)
```

```python
for link in soup.find_all('a', attrs={'href': re.compile("http")}):  
    soup_link = str(link)
    print(soup_link)
    file.write(soup_link)
file.flush() # to make the writing happen
file.close()  # to save the writing
```
