```python
from wordcloud import WordCloud, STOPWORDS
```

```python
with open("resume.txt", 'r') as file1:
    filedata = file1.read()
    
#txt must be encoded with UTF-8
```

```python
print(filedata[:200])
```

```python
stopwords = set(STOPWORDS)

wordcloud = WordCloud(stopwords=stopwords, max_words = 25, \
                     background_color="white").generate(filedata)

#max_words: calculate the 25 most popular words
#default background color is black
```

```python
import matplotlib.pyplot as mplib
mplib.imshow(wordcloud)
mplib.axis("off")
mplib.show()
```

```python
#to ignore some meaningless words
stopwords.update(["working","using"])
```

```python
wordcloud = WordCloud(stopwords=stopwords, max_words = 25, \
                     background_color="azure").generate(filedata)
mplib.imshow(wordcloud)
mplib.axis("off")
mplib.show()
```
